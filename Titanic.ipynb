## Titanic kazasında kimlerin hayatta kalacağını tahmin etmek.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split # Veriyi train-test olarak ayırır
from sklearn.linear_model import LogisticRegression # Makine öğrenmesi- tahmin modeli
from sklearn.metrics import accuracy_score # Modelin ne kadar doğru çalıştığını ölçer
```

```python
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
train.head()
# Verileri okuma
```

```python
# (NaN) değer varsa
# Age eksiklerini ortalama ile doldur
train['Age'] = train['Age'].fillna(train['Age'].mean())
test['Age']  = test['Age'].fillna(train['Age'].mean())

# Embarked eksiklerini en çok geçen değer ile doldur
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
test['Embarked']  = test['Embarked'].fillna(train['Embarked'].mode()[0])

# Fare eksiklerini ortalama ile doldur
test['Fare'] = test['Fare'].fillna(test['Fare'].mean())
```
